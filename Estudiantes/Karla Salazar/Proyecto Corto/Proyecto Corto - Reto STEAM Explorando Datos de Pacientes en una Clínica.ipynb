## Desaf칤o 1: An치lisis Cient칤fico - Reporte de Enfermedades Tratadas

### **Paso a Paso para el Estudiante**

#### **Paso 1: Crear las Listas**

### 

1. **Objetivo:** Necesitamos dos listas para almacenar:
    
    - Los nombres de las enfermedades.
    - La cantidad de veces que cada enfermedad aparece.
2. **Instrucci칩n:**
    
    - Crea una lista vac칤a llamada `lista_enfermedades`.
    - Crea otra lista vac칤a llamada `lista_cantidades`.

````python
# Lista para almacenar las enfermedades y sus cantidades
lista_enfermedades = []
lista_cantidades = []

````

#### **Paso 2: Recorrer los Datos de los Pacientes**

### 

1. **Objetivo:** Vamos a analizar cada paciente para extraer su enfermedad.
    
2. **Instrucci칩n:**
    
    - Usa un bucle `for` para recorrer cada `paciente` en `datos_pacientes` (ya est치 definida en el c칩digo como una lista de listas).
    - Dentro del bucle, asigna la enfermedad del paciente (칤ndice `4` de cada lista) a una variable llamada `enfermedad`.

````python
# Extraemos la enfermedad del paciente
#Puedes imprimir el for para ver los datos que se procesan

# Recorrer la lista de pacientes
for paciente in pacientes:
    # Extraer la enfermedad del paciente (칤ndice 4)
    enfermedad = paciente[4]

    # Imprimir la enfermedad de cada paciente
    print(f"Paciente con enfermedad: {enfermedad}")

````

#### **Paso 3: Verificar si la Enfermedad ya Est치 en la Lista**

1. **Objetivo:** Verificar si la enfermedad ya fue registrada.
    
2. **Instrucci칩n:**
    
    - Usa un `if` para verificar si la `enfermedad` ya est치 en `lista_enfermedades`.
    - Si **no** est치 en `lista_enfermedades`, agrega la enfermedad a `lista_enfermedades` y a침ade `1` a `lista_cantidades` para contarla por primera vez.
    - Si **s칤** est치 en `lista_enfermedades`, encuentra la posici칩n (`indice`) de la enfermedad en `lista_enfermedades` y aumenta en `1` el valor en `lista_cantidades` en la misma posici칩n.

````python
# Pista: Usa `in` para verificar si la enfermedad ya est치 en lista_enfermedades.
# Pista: Si la enfermedad no est치, agr칠gala a lista_enfermedades y empieza el conteo en lista_cantidades.
# Pista: Si la enfermedad ya est치, usa `index` para encontrar su posici칩n en lista_enfermedades y suma 1 en lista_cantidades en esa posici칩n.
# Recorrer la lista de pacientes para contar las enfermedades

for fila in pacientes:
    enfermedad = fila[4]  # Extraer la enfermedad (칤ndice 4)

    if enfermedad in lista_enfermedades:
        # Si la enfermedad ya est치 en la lista, aumentar su contador
        indice = lista_enfermedades.index(enfermedad)
        lista_cantidades[indice] += 1
    else:
        # Si la enfermedad no est치 en la lista, agregarla con cantidad 1
        lista_enfermedades.append(enfermedad)
        lista_cantidades.append(1)

````

#### **Paso 4: Imprimir el Reporte Final**

1. **Objetivo:** Mostrar el n칰mero de pacientes con cada enfermedad.
    
2. **Instrucci칩n:**
    
    - Usa un bucle `for` para recorrer la lista `lista_enfermedades`.
    - En cada iteraci칩n, imprime la enfermedad y el n칰mero de pacientes (que est치 en `lista_cantidades` en la misma posici칩n).

````python
print("\nReporte de enfermedades tratadas:")

for i in range(len(lista_enfermedades)):
    print(f"{lista_enfermedades[i]}: {lista_cantidades[i]} veces")
````

### **Resumen Completo del C칩digo**

En 칠sta celda resume el c칩digo completo que deber칤a tener al final para la soluci칩n del ejercicio:

````python
#Esta parte puedes colocar todos los bloques de las celdas anteriores 
import json

with open("clinica.json", "r", encoding="utf-8",errors ="replace") as archivo:
    pacientes = json.load(archivo)
    
# Lista para almacenar las enfermedades y sus cantidades
lista_enfermedades = []
lista_cantidades = []

for fila in pacientes:
    enfermedad = fila[4]  # Extraer la enfermedad (칤ndice 4)

    if enfermedad in lista_enfermedades:
        # Si la enfermedad ya est치 en la lista, aumentar su contador
        indice = lista_enfermedades.index(enfermedad)
        lista_cantidades[indice] += 1
    else:
        # Si la enfermedad no est치 en la lista, agregarla con cantidad 1
        lista_enfermedades.append(enfermedad)
        lista_cantidades.append(1)
        
print("\nReporte de enfermedades tratadas:")

for i in range(len(lista_enfermedades)):
    print(f"{lista_enfermedades[i]}: {lista_cantidades[i]} veces")
````

## Desaf칤o 3: Ingenier칤a - Comparar Pacientes por Identificaci칩n

Este desaf칤o te permitir치 comparar dos pacientes de la cl칤nica en base a sus identificaciones. Tu objetivo es averiguar:

1. **쯈u칠 enfermedades tienen en com칰n?**
2. **쯈u칠 medicamentos toman en com칰n?**

### Instrucciones:

1. Solicita dos identificaciones de pacientes al usuario.
2. Extrae las enfermedades y medicamentos de esos dos pacientes.
3. Compara las enfermedades y medicamentos y muestra los resultados.

### Preguntas:

- 쯈u칠 enfermedades y medicamentos son comunes en los pacientes comparados?
- 쯈u칠 informaci칩n adicional podr칤as incluir para mejorar esta comparaci칩n?


````python
# Pedimos las identificaciones de los dos pacientes

id1 = input("Ingrese la identificaci칩n del primer paciente: ")
id2 = input("Ingrese la identificaci칩n del segundo paciente: ")

# Funci칩n para obtener enfermedades y medicamentos de un paciente
def obtener_info_paciente(identificacion):
    enfermedades = []
    medicamentos = []

    for paciente in pacientes:
        if paciente[0] == identificacion:  # Comparar por ID
            if paciente[4] not in enfermedades:
                enfermedades.append(paciente[4])  # A침adir enfermedad si no est치 repetida
            if paciente[5] not in medicamentos:
                medicamentos.append(paciente[5])  # A침adir medicamento si no est치 repetido

    return enfermedades, medicamentos

# Obtener informaci칩n de los dos pacientes
enfermedades_p1, medicamentos_p1 = obtener_info_paciente(id1)
enfermedades_p2, medicamentos_p2 = obtener_info_paciente(id2)

# Encontrar enfermedades y medicamentos en com칰n
enfermedades_comunes = []
medicamentos_comunes = []

for enfermedad in enfermedades_p1:
    if enfermedad in enfermedades_p2:
        enfermedades_comunes.append(enfermedad)

for medicamento in medicamentos_p1:
    if medicamento in medicamentos_p2:
        medicamentos_comunes.append(medicamento)

# Mostrar los resultados
print("\nResultados de la comparaci칩n:")
#print(f"游댳 Enfermedades del paciente {id1}: {enfermedades_p1}")
#print(f"游댳 Enfermedades del paciente {id2}: {enfermedades_p2}")
print(f"Enfermedades en com칰n entre {id1} y {id2}: {enfermedades_comunes if enfermedades_comunes else 'Ninguna'}")

#print(f"\n游댳 Medicamentos del paciente {id1}: {medicamentos_p1}")
#print(f"游댳 Medicamentos del paciente {id2}: {medicamentos_p2}")
print(f"Medicamentos en com칰n entre {id1} y {id2}: {medicamentos_comunes if medicamentos_comunes else 'Ninguno'}")


"""
Prueba con 702370234 - 215487963
Salida esperada
Enfermedades en com칰n entre 702370234 y 215487963: ['migra침a']
Medicamentos en com칰n entre 702370234 y 215487963: ['acetaminofen']
"""
````

## Desaf칤o Final Opcional : Arte y Matem치ticas - Visualizaci칩n Creativa y Reflexi칩n

### Preguntas:

- 쯈u칠 patrones encontraste en los datos?
- 쮺칩mo podr칤as usar estos an치lisis para mejorar la atenci칩n de los pacientes?

````python
import json
import matplotlib.pyplot as plt
import pandas as pd

with open("clinica_intermedio.json", "r", encoding="utf-8", errors="replace") as archivo:
    pacientes = json.load(archivo)  # Se obtiene una lista de listas

# Extraer enfermedades y medicamentos
lista_enfermedades = [fila[4] for fila in pacientes]
lista_medicamentos = [fila[5] for fila in pacientes]

# Crear DataFrames con pandas
df_enfermedades = pd.DataFrame({'Enfermedad': lista_enfermedades})
df_medicamentos = pd.DataFrame({'Medicamento': lista_medicamentos})

# Contar ocurrencias
conteo_enfermedades = df_enfermedades['Enfermedad'].value_counts()
conteo_medicamentos = df_medicamentos['Medicamento'].value_counts()

# 游늼 Mostrar las tablas
print("\nFrecuencia de Enfermedades")
print(conteo_enfermedades.to_string())

print("\nFrecuencia de Medicamentos")
print(conteo_medicamentos.to_string())

# 游늵 Visualizaci칩n con gr치ficos
plt.figure(figsize=(10, 5))
conteo_enfermedades.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Frecuencia de Enfermedades en los Pacientes')
plt.xlabel('Enfermedades')
plt.ylabel('Cantidad de Pacientes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(8, 8))
conteo_medicamentos.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribuci칩n de Medicamentos Recetados')
plt.ylabel('')  # Oculta la etiqueta del eje y
plt.show()
````
